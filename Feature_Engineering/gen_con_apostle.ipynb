```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('event_time.csv', header=0, index_col=0)
apostles = pd.read_csv('apostleCallingYear.csv', header=0, index_col=0)
```

```python
data
```

```python
apostles.rename(columns={'Year':'Called_Year'}, inplace=True)
#there is an extra space in the index that needs to be removed
names = apostles.index
twelve = names.to_numpy()
twelve = np.array([person[:-1] for person in twelve])
apostles.set_index(twelve, inplace=True)
```

```python
total = data.merge(apostles, how='left', left_on='Speaker', right_index=True)
total['Called_Year'] = total['Called_Year'].fillna(value=np.inf)
```

```python
total
```

```python
is_apostle = [True if value >= call else False for value, call in zip(total['Year'], total['Called_Year'])]
total['Apostle'] = is_apostle
```

```python
total.drop(columns='Called_Year', inplace=True)
total
```
